#### BIVARIATE ANALYSIS ( Is gender has any impact on student's performance ? ) 

```python
def convert_gender_to_numeric(df, column_name='gender'):
    # Create a mapping of gender to numeric values
    gender_mapping = {'male': 0, 'female': 1}
    
    # Clean the gender column: convert to lowercase and strip spaces
    df[column_name] = df[column_name].str.strip().str.lower()
    
    # Replace the values in the gender column using the mapping
    df[column_name] = df[column_name].map(gender_mapping)
    
    # Handle any unmapped or NaN values by filling with a default value, e.g., -1
    df[column_name] = df[column_name].fillna(-1)
    
    # Return the modified DataFrame
    return df

df= convert_gender_to_numeric(df)
df
```
